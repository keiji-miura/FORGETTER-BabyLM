# Baby10M

# Figure 1

```python
import matplotlib.pyplot as plt; plt.figure(figsize=(8,4.12))
Layers =  [ 9, 12, 16, 18, 22,  24, 30, 36, 48]
N_heads = [16, 10,  8,  6,  8,   6,  4,  4,  3]

x = [x/10 for x in range(90,481)]
y = [120/x*10 for x in range(90,481)]
plt.plot(Layers, N_heads, "-ok", linewidth=3, label="Optimized Models")
plt.plot(x,  y, "--", label='"120-Rule"')
plt.xlabel('Number of Layers', fontsize=20)
plt.ylabel('Number of Heads per Layer', fontsize=20)
plt.tick_params(labelsize=18)
plt.yticks(range(2,17,2))
plt.grid()
plt.title("Baby10M, Normal", fontsize=16)
plt.legend(loc="upper right", fontsize=16, facecolor="white", framealpha=1)
```
